## Stationarity
Most time-series models assume that the underlying time-series data is **stationary**.

**Stationarity** is a statistical assumption that a time-series has:
*   **Constant mean**
*   **Constant variance**
*   **Autocovariance is independent or does not depend on time**

In other words, if we are using past data to predict future data, we should assume that the data will follow the same general trends and patterns as in the past. 

```python
def test_stationarity(data, column):
    """This function test the stationarity using moving average statistics and Dickey-Fuller test.
    
    Parameters
    ----------
    data: dataset
    column (str): column's name
    
    Returns
    -------
    plotted figures for  
    
    """
    
    # determing rolling statistics
    rolmean = data[column].rolling(window = 12, center = False).mean()
    rolstd = data[column].rolling(window = 12, center = False).std()
    
    # plot rolling statistics:
    orig = plt.plot(data[column], 
                    color = 'blue', 
                    label = 'Original')
    mean = plt.plot(rolmean, 
                    color = 'red', 
                    label = 'Rolling Mean')
    std = plt.plot(rolstd, 
                   color = 'black', 
                   label = 'Rolling Std')
    plt.legend(loc = 'best')
    plt.title('Rolling Mean & Standard Deviation for %s' %(column))
    plt.xticks(rotation = 45)
    plt.show(block = False)
    plt.close()
    
    # Perform Dickey-Fuller test:
    # Null Hypothesis (H_0): time series is not stationary
    # Alternate Hypothesis (H_1): time series is stationary
    print('Results of Dickey-Fuller Test:')
    df_test = adfuller(data[column], 
                      autolag='AIC')
    df_result = pd.Series(df_test[0:4], 
                         index = ['Test Statistic',
                                  'p-value',
                                  '# Lags Used',
                                  'Number of Observations Used'])
    for key, value in df_test[4].items():
        df_result['Critical Value (%s)'%key] = value
    print(df_result)
```

```python
test_stationarity(df, 'Price')
```

* With naked eyes, looking at the graph we can observe some  **non-stationary process**. To make sure, I run the Dickey-Fuller Test. Rolling mean looks like it changes over time. But standard deviation is mostly constant.
Based on Dickey-Fuller test, because p = 0.9967, we fail to reject the null hypothesis. That means the alternative hypothesis is not true and our time series is **not stationary**

### Dealing with non-stationarity
It is not uncommon for time series data to have to correct for non-stationarity.

There are 2 common reasons behind non-stationarity:

***Trend*** – mean is not constant over time. <br>
***Seasonality*** – variance is not constant over time.

From the plot, we may say that we need to deal more with Trend rather  Seasonality (Rolling Std is mostly constant)


If we do not correct for these things:

* Variance can be mis-specified
* Model fit can be worse
* Not leveraging valuable time-dependent nature of the data

---

### Eliminating trend and seasonality
*   **Transformation**
  *   *Example:* Log, square root, etc.
  -   We are going to look at log.
*   **Smoothing**
  *  *Example:* Weekly average, monthly average, rolling averages.
  -   We are going to look at weekly average.
*   **Differencing**
  *  *Example:* First-order differencing.
  -   We are going to look at first-order differencing.
*   **Polynomial Fitting**
  *  *Example:* Fit a regression model.
*   **Decomposition**

### Transformation, Smoothing, and Differencing
Applying log transformation, weekly moving average smoothing, and differencing are the ways to make data more stationary over time.

```python
def plot_transformed_data(data, col, col_transform):
  """
  This function plots transformed and original time series data
  """
  # plot time series data
  f, ax = plt.subplots(1,1)
  ax.plot(data[col])
  ax.plot(data[col_transform], color = 'red')

  # add title
  ax.set_title('%s and %s time-series graph' %(col, col_transform))

  # rotate x-labels
  ax.tick_params(axis = 'x', rotation = 45)

  # Add legend
  ax.legend([col, col_transform])
  
  plt.show()
  plt.close()
  
  return
```

```python
# transformation - log price
df['price_log'] = df['Price'].apply(lambda x: np.log(x))

# transformation - 7-day moving averages of log price
df['price_log_moving_avg'] = df['price_log'].rolling(window = 7, center = False).mean()

# transformation - 7-day moving average of price
df['price_moving_avg'] = df['Price'].rolling(window = 7, center = False).mean()

# transformation - Difference between logged price and first-order difference logged price
df['price_log_diff'] = df['price_log'].diff()

# transformation - Difference between price and moving average price
df['price_moving_avg_diff'] = df['Price'] - df['price_moving_avg']


# drop missing values
df_transform = df.dropna()

# display data
display(df_transform.head())

# plot price_log
plot_transformed_data(df_transform, 'Price', 'price_log')

# plot price_log_moving_avg
plot_transformed_data(df_transform, 'Price', 'price_log_moving_avg')

# plot price_moving_avg
plot_transformed_data(df_transform, 'Price', 'price_moving_avg')

# plot price_log_diff
plot_transformed_data(df_transform, 'Price', 'price_log_diff')

# plot price_moving_avg_diff
plot_transformed_data(df_transform, 'Price', 'price_moving_avg_diff')
```

```python
# stationarity test for price_log
test_stationarity(df_transform, 'price_log')

# stationarity test for price_moving_avg
test_stationarity(df_transform, 'price_moving_avg')

# stationarity test for price_log
test_stationarity(df_transform, 'price_log_moving_avg')

# stationarity test for price_log
test_stationarity(df_transform, 'price_log_diff')

# stationarity test for price_log
test_stationarity(df_transform, 'price_moving_avg_diff')
```

### Decomposition: Trend, Seasonality, Residuals

```python
def plot_decomposition(data, col, trend, seasonal, residual):
  """
  This function plots time series original data, Trends, Seasonality and Residuals
  
  """
  f, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2, figsize = (15, 5), sharex = True)

  ax1.plot(data[col], label = 'Original')
  ax1.legend(loc = 'best')
  ax1.tick_params(axis = 'x', rotation = 45)

  ax2.plot(data[trend], label = 'Trend')
  ax2.legend(loc = 'best')
  ax2.tick_params(axis = 'x', rotation = 45)

  ax3.plot(data[seasonal],label = 'Seasonality')
  ax3.legend(loc = 'best')
  ax3.tick_params(axis = 'x', rotation = 45)

  ax4.plot(data[residual], label = 'Residuals')
  ax4.legend(loc = 'best')
  ax4.tick_params(axis = 'x', rotation = 45)
  plt.tight_layout()

  plt.suptitle('Trend, Seasonal, and Residual Decomposition of %s' %(col), 
               x = 0.5, 
               y = 1.05, 
               fontsize = 18)
  plt.show()
  plt.close()
  
  return
```

```python
decomposition = seasonal_decompose(df_transform['price_log'], freq = 365)

df_transform.loc[:,'trend'] = decomposition.trend
df_transform.loc[:,'seasonal'] = decomposition.seasonal
df_transform.loc[:,'residual'] = decomposition.resid

# plot decomposition
plot_decomposition(df_transform, 'price_log', 'trend', 'seasonal', 'residual') 
                   
# test for Residual stationarity
test_stationarity(df_transform.dropna(), 'residual')
```

# ARIMA

---
We can use ARIMA models when we know there is dependence between values and we can leverage that information to forecast.

ARIMA = Auto-Regressive Integrated Moving Average. <br>
***Assumptions:*** The time-series is stationary. <br>
***Depends on:***
1. Number of AR (Auto-Regressive) terms (p).
2. Number of I (Integrated or Difference) terms (d).
3. Number of MA (Moving Average) terms (q).

---

To determine above mentioned p, d, q, I'll use ACF (Autocorrelation Function) & PACF (Partial Autocorrelation Function) plots. <br>
* **ACF:**  Correlation between the time series with a lagged version of itself. 
* **PACF:** Additional correlation explained by each successive lagged term.



```python
# calculate ACF and PACF
lag_acf = acf(np.array(df_transform['price_log_diff']), nlags = 20)
lag_pacf = pacf(np.array(df_transform['price_log_diff']), nlags = 20)

# set up subplots
f, (ax1, ax2) = plt.subplots(1,2, figsize = (10, 5)) 

# plot ACF: 

ax1.plot(lag_acf)
ax1.axhline(y=0,linestyle='--',color='red')
ax1.axhline(y=-1.96/np.sqrt(len(df_transform['price_log_diff'])),linestyle='--',color='gray')
ax1.axhline(y=1.96/np.sqrt(len(df_transform['price_log_diff'])),linestyle='--',color='gray')
ax1.set_title('Autocorrelation Function for %s' %('price_log_diff'))

# plot PACF:
ax2.plot(lag_pacf)
ax2.axhline(y=0,linestyle='--',color='red')
ax2.axhline(y=-1.96/np.sqrt(len(df_transform['price_log_diff'])),linestyle='--',color='gray')
ax2.axhline(y=1.96/np.sqrt(len(df_transform['price_log_diff'])),linestyle='--',color='gray')
ax2.set_title('Partial Autocorrelation Function for %s' %('price_log_diff'))
  
plt.tight_layout()
plt.show()
plt.close()
```

*  What actually **p** and **q** values are? <br>
p – Lag value where the PACF chart crosses the upper confidence interval for the first time. <br>
q – Lag value where the ACF chart crosses the upper confidence interval for the first time.

From the figures, we can observe that the p=q

```python
def run_arima(data, col, p, d, q):
  """ This function runs ARIMA model and returns results as graphs
  
  Parameters
    ----------
    data: dataset
    col (str): time-series column
    p: number of AR terms
    d: number of I terms
    q: number of MA terms
    
    Returns
    -------
    plotted figures for ARIMA model  
    
    """
  

  # fit ARIMA model on time series
  model = ARIMA(data[col], order=(p, d, q))  
  results = model.fit(disp=-1)  
  
  # get lengths correct to calculate RSS
  len_results = len(results.fittedvalues)
  col_modified = data[col][-len_results:]
  
  # calculate root mean square error (RMSE) and residual sum of squares (RSS)
  rss = sum((results.fittedvalues - col_modified)**2)
  rmse = np.sqrt(rss / len(data[col]))
  
  # plot fit
  plt.plot(data[col])
  plt.plot(results.fittedvalues, color = 'blue')
  plt.title('For ARIMA model (%i, %i, %i) for col %s, RSS: %.4f, RMSE: %.4f' %(p, d, q, col, rss, rmse))
  
  plt.show()
  plt.close()
  
  return results
```

# Facebook Prophet Model

---

Prophet is an additive regression model that includes highly advanced, intelligent **forecasting methods**. What can we do with it?

* Hourly, daily, or weekly observations with at least a few months of history
* Historical trend changes due to product launches or logging changes
* Automatically detect changes in trends by selecting ***changepoints***

**Note:** this model is designed to work on a daily level. It still works with other inputs very good.

```python
from fbprophet import Prophet
import datetime
from datetime import datetime
```

```python
# function to calculate the number of days between two dates
def days_between(d1, d2):
    """Calculate the number of days between two dates.  D1 is start date (inclusive) and d2 is end date (inclusive)"""
    d1 = datetime.strptime(d1, "%Y-%m-%d")
    d2 = datetime.strptime(d2, "%Y-%m-%d")
    return abs((d2 - d1).days + 1)
```

```python
# define parameters for model input
start_forecasting_date = '2018-01-01'
end_forecasting_date = '2020-12-31'

future_num_points = days_between(start_forecasting_date, end_forecasting_date)
```

```python
# get relevant data
df_prophet = df_transform[['Price']] 

# reset index
df_prophet = df_prophet.reset_index()

# rename columns
df_prophet = df_prophet.rename(columns = {'Date': 'ds', 'Price': 'y'}) 

# change 'ds' type from datetime to date (necessary for FB Prophet)
df_prophet['ds'] = pd.to_datetime(df_prophet['ds'])

# change 'y' type to numeric (necessary for FB Prophet)
df_prophet['y'] = pd.to_numeric(df_prophet['y'], errors='ignore')
```

```python
def run_daily_forecast(data,
                       holidays,
                       growth,
                       n_changepoints = 25,
                       changepoint_prior_scale = 0.05,
                       changepoints = None,
                       holidays_prior_scale = 10,
                       interval_width = 0.8,
                       mcmc_samples = 1,
                       future_num_points = 10, 
                       daily_seasonality = True):
  """
  This function creates daily forecast
  
  """
  
  # create copy of dataframe
  data_ = data.copy()

  # initialize the model
  m = Prophet(growth = growth,
              n_changepoints = n_changepoints,
              changepoint_prior_scale = changepoint_prior_scale,
              changepoints = changepoints,
              holidays = holidays,
              holidays_prior_scale = holidays_prior_scale,
              interval_width = interval_width,
              mcmc_samples = mcmc_samples, 
              daily_seasonality = daily_seasonality)

  # fit model
  m.fit(data_)

  # dataframe for predictions
  future = m.make_future_dataframe(periods = future_num_points)
  
  # Create predictions
  forecast = m.predict(future)
  
  

  # Plot
  m.plot(forecast);
  m.plot_components(forecast)

  return forecast
```
